# Analyse Exploratoire

### Import des modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import missingno as msno
```

### Analyse Exploratoire

```python
building_consumption = pd.read_csv("2016_Building_Energy_Benchmarking.csv")

target_energy = 'SiteEnergyUse(kBtu)'
target_emissions = 'TotalGHGEmissions'

donnees_structurelles = [
    'OSEBuildingID', 'BuildingType', 'PrimaryPropertyType', 'ZipCode', 'CouncilDistrictCode',
    'Neighborhood', 'Latitude', 'Longitude', 'YearBuilt', 'NumberofBuildings', 'NumberofFloors',
    'PropertyGFATotal', 'PropertyGFAParking', 'PropertyGFABuilding(s)', 'ListOfAllPropertyUseTypes',
    'LargestPropertyUseType', 'LargestPropertyUseTypeGFA', 'SecondLargestPropertyUseType',
    'SecondLargestPropertyUseTypeGFA', 'ThirdLargestPropertyUseType', 'ThirdLargestPropertyUseTypeGFA'
]
non_structurelles = [col for col in building_consumption.columns if col not in donnees_structurelles]

# On regarde comment un batiment est dfini dans ce jeu de donnes 
building_consumption.head()

msno.matrix(building_consumption.sample(250))
missing_values = building_consumption.isnull().mean().sort_values(ascending=False)
print("\nValeurs manquantes par colonne (%):\n", (missing_values[missing_values > 0] * 100).round(2))
```

```python
classification_building_consumption = pd.DataFrame({
    "Colonnes structurelles": pd.Series(donnees_structurelles),
    "Colonnes non structurelles": pd.Series(non_structurelles)
})

classification_building_consumption
```

```python
from matplotlib.patches import Patch

#  Compter les occurrences de chaque catgorie dans BuildingType
building_type_counts = building_consumption['BuildingType'].value_counts()
building_type_order = building_type_counts.index.tolist()

# Dfinir les couleurs en fonction de la prsence de "nonresidential"
bar_colors = [
    'orange' if isinstance(bt, str) and 'nonresidential' in bt.lower() else 'blue'
    for bt in building_type_order
]

# Afficher le graphique
plt.figure(figsize=(10, 6))
building_type_counts.sort_values().plot.barh(
    color=[bar_colors[building_type_order.index(bt)] for bt in building_type_counts.sort_values().index]
)

plt.title("Rpartition des types de btiments")
plt.xlabel("Nombre de btiments")
plt.ylabel("Type de btiment")

# Ajouter la lgende en bas  droite
legend_elements = [
    Patch(facecolor='orange', label='NonResidential'),
    Patch(facecolor='blue', label='Autres')
]
plt.legend(handles=legend_elements, title="Catgories", loc='lower right')

plt.tight_layout()
plt.show()

# Afficher le nombre de lignes avant filtrage
print("Nombre de lignes avant filtrage:", building_consumption.shape[0])
# 2 Filtrer uniquement les btiments NonResidential
building_consumption = building_consumption[building_consumption['BuildingType'].str.contains("nonresidential", case=False, na=False)]
print("Nombre de lignes aprs filtrage (seulement NonResidential):", building_consumption.shape[0])
```

```python
cols_utiles = [
    'SiteEnergyUse(kBtu)', 'TotalGHGEmissions', 'ENERGYSTARScore',
    'YearBuilt', 'NumberofFloors', 'PropertyGFATotal', 'Outlier', 'ComplianceStatus'
]
df_clean = building_consumption[cols_utiles].copy()

# Histogrammes pour visualiser les distributions
fig, axes = plt.subplots(3, 2, figsize=(14, 12))
df_clean['SiteEnergyUse(kBtu)'].hist(ax=axes[0, 0], bins=50)
axes[0, 0].set_title("SiteEnergyUse(kBtu)")

df_clean['TotalGHGEmissions'].hist(ax=axes[0, 1], bins=50)
axes[0, 1].set_title("TotalGHGEmissions")

df_clean['ENERGYSTARScore'].hist(ax=axes[1, 0], bins=25)
axes[1, 0].set_title("ENERGYSTARScore")

df_clean['YearBuilt'].hist(ax=axes[1, 1], bins=50)
axes[1, 1].set_title("YearBuilt")

df_clean['NumberofFloors'].hist(ax=axes[2, 0], bins=30)
axes[2, 0].set_title("NumberofFloors")

df_clean['PropertyGFATotal'].hist(ax=axes[2, 1], bins=50)
axes[2, 1].set_title("PropertyGFATotal")

plt.tight_layout()
plt.show()

# Aperu des outliers ou valeurs aberrantes
print("\nValeurs aberrantes dtectes :")
print("Anne de construction < 1800 ou > 2025 :")
print(building_consumption[(building_consumption['YearBuilt'] < 1800) | (building_consumption['YearBuilt'] > 2025)][['YearBuilt']])

print("\nBtiments avec 0 tage ou > 100 tages :")
print(building_consumption[(building_consumption['NumberofFloors'] <= 0) | (building_consumption['NumberofFloors'] > 100)][['NumberofFloors']])

print("\nConsommation ou mission nulle :")
print(building_consumption[(building_consumption['SiteEnergyUse(kBtu)'] <= 0) | (building_consumption['TotalGHGEmissions'] <= 0)][['SiteEnergyUse(kBtu)', 'TotalGHGEmissions']])

print("\nBtiments marqus comme outliers :")
print(building_consumption[building_consumption['Outlier'].notna()][['Outlier']].value_counts())

# Supprimer les valeurs aberrantes
conditions_aberrantes = (
    (building_consumption['YearBuilt'] < 1800) | (building_consumption['YearBuilt'] > 2025) |
    (building_consumption['NumberofFloors'] <= 0) | (building_consumption['NumberofFloors'] > 100) |
    (building_consumption['SiteEnergyUse(kBtu)'] <= 0) |
    (building_consumption['TotalGHGEmissions'] < 0)
)
nb_avant = building_consumption.shape[0]
building_consumption = building_consumption[~conditions_aberrantes].copy()
nb_apres = building_consumption.shape[0]

# Rsum du nettoyage
print(f" Lignes avant nettoyage : {nb_avant}")
print(f" Lignes aprs nettoyage : {nb_apres}")
print(f" Lignes supprimes : {nb_avant - nb_apres}")
```

```python
cols_to_check = [
    'PropertyGFABuilding(s)', 
    'LargestPropertyUseTypeGFA', 
    'SecondLargestPropertyUseTypeGFA', 
    'ThirdLargestPropertyUseTypeGFA'
]

# 3 Extraire les colonnes concernes
df_use = building_consumption[cols_to_check].copy()

# 4 Calcul de la somme des usages dclars (en remplaant les NaN par 0)
df_use['Somme_GFAs'] = df_use[
    ['LargestPropertyUseTypeGFA', 'SecondLargestPropertyUseTypeGFA', 'ThirdLargestPropertyUseTypeGFA']
].fillna(0).sum(axis=1)

# 5 Calcul de l'cart entre surface dclare et somme des usages
df_use['Ecart'] = df_use['PropertyGFABuilding(s)'] - df_use['Somme_GFAs']

# 6 Affichage des graphes
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

#  Scatter plot : Surface dclare vs Somme des usages
axes[0].scatter(df_use['PropertyGFABuilding(s)'], df_use['Somme_GFAs'], alpha=0.5)
axes[0].plot([0, df_use['PropertyGFABuilding(s)'].max()],
             [0, df_use['PropertyGFABuilding(s)'].max()], 'r--')
axes[0].set_xlabel("Surface totale dclare (PropertyGFABuilding(s))")
axes[0].set_ylabel("Somme des usages GFA")
axes[0].set_title("Comparaison Surface Totale vs Somme des Usages")

#  Histogramme des carts
axes[1].hist(df_use['Ecart'].dropna(), bins=50)
axes[1].set_xlabel("cart (surface dclare - somme usages)")
axes[1].set_ylabel("Nombre de btiments")
axes[1].set_title("Distribution des carts de surface")

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Supprimer les colonnes constantes
df_structurelles = building_consumption[donnees_structurelles].copy()
df_non_structurelles = building_consumption[non_structurelles].copy()

df_structurelles = df_structurelles.loc[:, df_structurelles.nunique() > 1]
df_non_structurelles = df_non_structurelles.loc[:, df_non_structurelles.nunique() > 1]

# LabelEncoder pour les colonnes catgorielles
label_encoder = LabelEncoder()

for col in df_structurelles.select_dtypes(include='object').columns:
    df_structurelles[col] = label_encoder.fit_transform(df_structurelles[col].astype(str))

for col in df_non_structurelles.select_dtypes(include='object').columns:
    df_non_structurelles[col] = label_encoder.fit_transform(df_non_structurelles[col].astype(str))

# Remplir les NaN par la moyenne (ou mdiane si tu prfres)
df_structurelles = df_structurelles.fillna(df_structurelles.mean(numeric_only=True))
df_non_structurelles = df_non_structurelles.fillna(df_non_structurelles.mean(numeric_only=True))

# Normalisation
scaler = StandardScaler()
df_structurelles_scaled = pd.DataFrame(scaler.fit_transform(df_structurelles), columns=df_structurelles.columns)
df_non_structurelles_scaled = pd.DataFrame(scaler.fit_transform(df_non_structurelles), columns=df_non_structurelles.columns)

# Calcul de la corrlation croise
correlation_matrix = pd.DataFrame(index=df_structurelles_scaled.columns, columns=df_non_structurelles_scaled.columns)

for struct_col in df_structurelles_scaled.columns:
    for non_struct_col in df_non_structurelles_scaled.columns:
        correlation_matrix.loc[struct_col, non_struct_col] = df_structurelles_scaled[struct_col].corr(df_non_structurelles_scaled[non_struct_col])

correlation_matrix = correlation_matrix.astype(float)

# Visualisation
plt.figure(figsize=(16, 10))
sns.heatmap(
    correlation_matrix,
    cmap='coolwarm',
    center=0,
    annot=True,                #  Active les annotations
    fmt=".2f",                 #  Format des nombres (2 dcimales)
    linewidths=0.5,            #  Sparation visuelle entre cases
    cbar_kws={'label': 'Corrlation'}
)
plt.title("Corrlation normalise entre variables structurelles et non structurelles")
plt.xlabel("Variables non structurelles")
plt.ylabel("Variables structurelles")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
